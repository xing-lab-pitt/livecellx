```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from livecellx.preprocess.utils import normalize_img_to_uint8, enhance_contrast
# Load the new video
# new_video_path = './resources/hela-st-wrong.mp4'

# data fig
# new_video_path = './resources/cxa-drop-div-test_mitosis_35_combined_padding-100.mp4'
# new_video_path = './resources/test_mitosis_16_combined_padding-20.mp4'
# new_video_path = './resources/test_mitosis_16_combined_padding-200.mp4'
new_video_path = './resources/test_mitosis_17_combined_padding-200.mp4'
new_cap = cv2.VideoCapture(new_video_path)

# Check if video loaded successfully
if not new_cap.isOpened():
    raise IOError("Error: Could not open new video.")

# Read all frames from the new video
new_frames = []
while True:
    ret, frame = new_cap.read()
    if not ret:
        break
    new_frames.append(frame)

new_cap.release()


# Display all frames from the new video in a single row without any margins and separate the RGB channels
# fig, axes = plt.subplots(3, len(new_frames), figsize=(8, 4))  # 3 rows for R, G, B channels
fig, axes = plt.subplots(3, len(new_frames), figsize=(16, 8))  # 3 rows for R, G, B channels

# Remove margins between subplots
plt.subplots_adjust(wspace=0, hspace=0)

for i, frame in enumerate(new_frames):
    # Split the frame into R, G, B channels
    R, G, B = cv2.split(frame)
    R = enhance_contrast(R, factor=3)
    # R = normalize_img_to_uint8(R)
    # Display the R channel
    # axes[0, i].imshow(R, cmap='Grays')
    axes[0, i].imshow(R, cmap='Greys')
    axes[0, i].axis('off')

    # Display the G channel
    axes[1, i].imshow(G, cmap='inferno')
    axes[1, i].axis('off')

    # Display the B channel
    axes[2, i].imshow(B, cmap='Blues')
    axes[2, i].axis('off')

plt.subplots_adjust(wspace=0, hspace=0)
# Adjust layout to minimize space
plt.tight_layout(pad=0)

plt.show()
```

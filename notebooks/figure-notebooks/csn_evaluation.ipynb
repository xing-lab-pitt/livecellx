```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = sns.load_dataset("penguins")
```

```python
df[:2]
```

```python
data = [
    ["origin", "iou", "test", 0.5],
    ["origin", "iou", "train", 0.52],
    ["origin", "accuracy", "test", 0.95],
    ["origin", "accuracy", "train", 0.95],

    ["mild-aug", "iou", "test", 0.74],
    ["mild-aug", "iou", "train", 0.90],
    ["mild-aug", "accuracy", "test", 0.99],
    ["mild-aug", "accuracy", "train", 0.99],

    ["strong-aug", "iou", "test", 0.68],
    ["strong-aug", "iou", "train", 0.70],
    ["strong-aug", "accuracy", "test", 0.98],
    ["strong-aug", "accuracy", "train", 0.98],

    ["no-aug", "iou", "test", 0.74],
    ["no-aug", "iou", "train", 0.89],
    ["no-aug", "accuracy", "test", 0.99],
    ["no-aug", "accuracy", "train", 0.99],
]
```

```python
df = pd.DataFrame(data, columns=["model", "metric", "split", "value"])

train_df = df[df["split"] == "train"]
sns.set_style('darkgrid')
figure = plt.figure(figsize=(10, 7), dpi=400)
sns.barplot(x="metric", y="value", hue="model", data=train_df, width=0.5, orient="v")
plt.title("$\\bf{Train\ Set\ Performance}$",  fontsize=20)
plt.xlabel("Metric", fontsize=20)
plt.ylabel("Value", fontsize=20)

test_df = df[df["split"] == "test"]
sns.set_style('darkgrid')
figure = plt.figure(figsize=(10, 7), dpi=400)
sns.barplot(x="metric", y="value", hue="model", data=test_df, width=0.5, orient="v")
plt.title("$\\bf{Test\ Set\ Performance}$",  fontsize=20)
plt.xlabel("Metric", fontsize=20)
plt.ylabel("Value", fontsize=20)
```

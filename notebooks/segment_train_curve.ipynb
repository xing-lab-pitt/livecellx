```python
import seaborn as sns
from pathlib import Path
import json
import matplotlib.pyplot as plt



def plot_train_curve(metric_path, **kwargs):
    metric_data = []
    with open(metric_path, "r") as f:
        for line in f:
            metric_data.append(json.loads(line))

    metric_data[0].keys()
    total_loss = [x["total_loss"] for x in metric_data]
    plt.plot(total_loss, **kwargs)

output_path = r"D:\LiveCellTracker-dev\notebooks\notebook_results\train_log\detectron_train_output__ver0.0.1_2_experts"
metric_path = Path(output_path) / "metrics.json"
plot_train_curve(metric_path, label="3-expert")

output_path = r"D:\LiveCellTracker-dev\notebooks\notebook_results\train_log\detectron_train_output__ver0.0.2"
metric_path = Path(output_path) / "metrics.json"
plot_train_curve(metric_path, label="2-expert")

output_path = r"D:\LiveCellTracker-dev\notebooks\notebook_results\train_log\detectron_run3_output"
metric_path = Path(output_path) / "metrics.json"
plot_train_curve(metric_path, label="before_correct")

plt.title("Training loss curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
```
